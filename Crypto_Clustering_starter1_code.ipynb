```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# Normalize the data using StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
# Copy the crypto names from the original DataFrame
# Set the coin_id column as index
df_market_data_scaled = pd.DataFrame(
    scaled_data,
    columns=df_market_data.columns,
    index=df_market_data.index
)


# Display the scaled DataFrame
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 12)
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(df_market_data_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve = df_elbow.hvplot.line( 
    x='k', 
    y='inertia', 
    line_width=2, 
    title='Elbow Curve to Identify Optimal k', 
    xlabel='Number of Clusters (k)', 
    ylabel='Inertia', 
    grid=True ) 
# Display the plot 
hvplot.show(elbow_curve)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
optimal_k = 4
kmeans_model = KMeans(n_clusters=optimal_k, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
clusters = kmeans_model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
print("Cluster Values:", clusters)
```

```python
# Create a copy of the scaled DataFrame
df_clustered = df_market_data_scaled.copy()
df_clustered['Cluster'] = clusters
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_clustered['Cluster'] = clusters
# Display the copy of the scaled DataFrame
print("Clustered DataFrame:")
print(df_clustered.head())
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = df_clustered.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="Cluster",
    colormap='viridis',
    hover_cols=['Cluster'],
    title="Cryptocurrency Clustering",
    xlabel="24h Price Change (%)",
    ylabel="7d Price Change (%)"
)

hvplot.show(scatter_plot)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_market_data_scaled)

# View the scaled PCA data
explained_variance = pca.explained_variance_ratio_
total_explained_variance = explained_variance.sum()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
df_pca = pd.DataFrame(
    pca_data,
    columns=['PC1', 'PC2', 'PC3'],
    index=df_market_data_scaled.index
)

# Add the cluster labels to the PCA DataFrame
df_pca['Cluster'] = clusters

# Outputs
scatter_plot, total_explained_variance, df_pca.head()
```

```python
Question: After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

 

Using fewer features to cluster the data using K-Means, especially through techniques like PCA (Principal Component Analysis), can have several impacts:

Benefits:

Reduced Dimensionality: Simplifies the data by reducing the number of features while retaining most of the original information. This makes the clustering process faster and computationally efficient.
Improved Visualization: With fewer dimensions, it's easier to visualize the clusters in 2D or 3D plots, which helps in understanding the cluster structure and relationships.
Noise Reduction: PCA helps in removing noise from the data by focusing on the most significant features, which can lead to more distinct and meaningful clusters.
Avoiding Overfitting: By reducing the number of features, the model is less likely to overfit, thus potentially improving the generalizability of the clustering results.
Drawbacks:

Loss of Information: While PCA tries to retain as much information as possible, some information is inevitably lost when reducing dimensions, which could potentially lead to less accurate clustering.
Interpretability: The principal components generated by PCA are linear combinations of the original features, which can make it harder to interpret the meaning of each component and how they contribute to the clusters.
Dependence on Variance: PCA focuses on maximizing variance, which means features that do not vary much but are still important for clustering might be underrepresented in the principal components.
Analysis:

With PCA (Fewer Features): Clusters might appear more distinct and cleaner in the visualizations, and the clustering process might be faster. However, you might need to understand how the principal components relate to the original features to interpret the results properly.
Without PCA (More Features): Clusters might be based on more detailed information, potentially leading to more accurate but complex clustering results. Visualizing high-dimensional data is challenging, which might make understanding cluster relationships harder.
In essence, using fewer features via PCA can streamline the clustering process and make the results more interpretable in lower dimensions, but it requires careful consideration of the trade-off between simplicity and the loss of potentially important information.
```
